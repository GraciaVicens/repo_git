```python
from pandas import DatetimeIndex
```

¿Cuántas veces ha nevado por año (snowfall)?

```python
df = pd.read_csv("data/rdu-weather-history.csv",sep=";")
df.index = DatetimeIndex(df["date"])
df.head()
```

```python
df2 = df[df["snowfall"] > 0]
nevadas = df2.groupby(df2.index.year).agg({'snowfall': 'count'})
nevadas
```

¿En qué año se han registrado más nieve (snowdepth)?

```python
valor_maximo = df["snowdepth"].max()
anio_max = df[df[ "snowdepth"] == valor_maximo] 
anio= anio_max.index.year[0]
print("En el año: ", anio)
```
